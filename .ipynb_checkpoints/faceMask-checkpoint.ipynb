```python
import os
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import xmltodict
import random
import tensorflow as tf
from xml.dom import minidom

import tensorflow_datasets as tfds
import tensorflow as tf
```

<h1>Loading the data (images and labels)</h1>

<h2>Loading images</h2>

```python
def read_image(path, filename):
    img = cv2.imread(os.path.join(path, filename))
    if img is not None:
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        return img
```

```python
def load_images(path):
    images = []
    for filename in sorted(os.listdir(path)):
        images.append(read_image(path, filename))
    return images
```

<h2>Loading labels</h2>

```python
def parse_xml(path):
    mydoc = minidom.parse(path)
    img_size = {}
    img_size["width"] = int(mydoc.getElementsByTagName("width")[0].firstChild.data)
    img_size["height"] = int(mydoc.getElementsByTagName("height")[0].firstChild.data)
    
    # start reading the image labels
    xml_items = ["xmin", "ymin", "xmax", "ymax", "name"]
    labels = []
    img_label = {}
    for obj in mydoc.getElementsByTagName("object"):
        for item in xml_items:
            if item == "name":
                img_label["name"] = obj.getElementsByTagName(item)[0].firstChild.data
            else:
                img_label[item] = int(obj.getElementsByTagName(item)[0].firstChild.data)
        labels.append(img_label.copy())
    
    return labels, img_size
```

```python
def load_labels(path):
    all_labels = []
    for filename in sorted(os.listdir(path)):
        lb = {}
        labels, img_size = parse_xml(os.path.join(path, filename))
        lb["labels"] = labels
        lb["size"] = img_size
        all_labels.append(lb)
    return all_labels
```

<h3>loading the dataset</h3>

```python
images = load_images("dataset/images")
print(len(images))
```

```python
labels = load_labels("dataset/annotations")
print(len(labels))
```

```python
def rectangle_location(label):
    xy_min = (label["xmin"], label["ymin"])
    xy_max = (label["xmax"], label["ymax"])
    return xy_min, xy_max
```

<h2>Visualizing image with labels</h2>

```python
def visualize_image(img, labels, img_size):
    thickness = int((img_size["width"]+img_size["height"])/400.)
    for label in labels:
        xy_min, xy_max = rectangle_location(label)
        if label["name"] == 'with_mask':
            cv2.rectangle(img, xy_min, xy_max, (0, 255, 0), thickness)
        elif label["name"] == 'without_mask':
            cv2.rectangle(img, xy_min, xy_max, (255, 0, 0), thickness)
        else: 
            cv2.rectangle(img, xy_min, xy_max, (0, 0, 255), thickness)
    
    plt.figure(figsize=(20, 20))
    plt.subplot(1, 2, 1)
    plt.axis('off')
    plt.imshow(img)
    plt.show()
    
```

```python
index = 0
visualize_image(images[index], labels[index]["labels"], labels[index]["size"])
```

<h1>Data pre-processing</h1>

```python
def crop_faces_from_image(img, labels):
    croped_images_labels = []
    for label in labels:
        xy_min, xy_max = rectangle_location(label)
        croped_face = img[xy_min[1]:xy_max[1], xy_min[0]:xy_max[0]]
        without_mask = 0
        if label["name"] == "without_mask":
            without_mask = 1
        croped_image_label = [croped_face, without_mask]
        croped_images_labels.append(croped_image_label)
    
    return croped_images_labels
```

<h2>Saving faces of the images in folders</h2>

```python
with_mask_dir = "training_dataset/with_mask/"
without_mask_dir = "training_dataset/without_mask/"

with_mask_counter = 0
without_mask_counter = 0

for image, label in zip(images, labels):
    croped_images_labels = crop_faces_from_image(image, label["labels"])
    for croped_image_label in croped_images_labels:
        img = croped_image_label[0]
        lb = croped_image_label[1]      
        if lb == 0:
            cv2.imwrite(with_mask_dir + str(with_mask_counter) + ".jpg", img)
            with_mask_counter += 1
        elif lb == 1:
            cv2.imwrite(without_mask_dir + str(without_mask_counter) + ".jpg", img)
            without_mask_counter += 1
```

```python
def number_of_sample():
    num_samples_with_mask = len(os.listdir("training_dataset/with_mask"))
    num_samples_without_mask = len(os.listdir("training_dataset/without_mask"))
    
    total_samples = num_samples_with_mask + num_samples_without_mask
    
    print("Total number of samples = ", total_samples)
    print("Number of samples with mask = ", num_samples_with_mask)
    print("Number of samples without mask = ", num_samples_without_mask)
```

<h1>TensorFlow: reading images</h1>

```python
# builder = tfds.ImageFolder('training_dataset/')
```

```python
batch_size = 32
img_height = 224
img_width = 224

train_dataset = tf.keras.preprocessing.image_dataset_from_directory(
    "training_dataset",
    shuffle=True,
    seed=42,
    validation_split=0.3,
    subset="training",
    image_size=(img_height, img_width),
    batch_size=batch_size)

validation_dataset = tf.keras.preprocessing.image_dataset_from_directory(
    "training_dataset",
    shuffle=True,    
    seed=42,
    validation_split=0.3,
    subset="validation",
    image_size=(img_height, img_width),
    batch_size=batch_size)
```

```python
val_batches = tf.data.experimental.cardinality(validation_dataset)
test_dataset = validation_dataset.take(val_batches // 5)
validation_dataset = validation_dataset.skip(val_batches // 5)
```

```python
print('Number of train batches: %d' % tf.data.experimental.cardinality(train_dataset))
print('Number of validation batches: %d' % tf.data.experimental.cardinality(validation_dataset))
print('Number of test batches: %d' % tf.data.experimental.cardinality(test_dataset))
```

<h2>Data pre-processing</h2>

```python
AUTOTUNE = tf.data.experimental.AUTOTUNE

train_dataset = train_dataset.prefetch(buffer_size=AUTOTUNE)
validation_dataset = validation_dataset.prefetch(buffer_size=AUTOTUNE)
test_dataset = test_dataset.prefetch(buffer_size=AUTOTUNE)
```

```python
data_augmentation = tf.keras.Sequential([
  tf.keras.layers.experimental.preprocessing.RandomFlip('horizontal'),
  tf.keras.layers.experimental.preprocessing.RandomRotation(0.2),
])
```

```python
for image, _ in train_dataset.take(1):
    plt.figure(figsize=(10, 10))
    first_image = image[0]
    for i in range(9):
        ax = plt.subplot(3, 3, i + 1)
        augmented_image = data_augmentation(tf.expand_dims(first_image, 0))
        plt.imshow(augmented_image[0] / 255)
        plt.axis('off')
```

```python
preprocess_input = tf.keras.applications.mobilenet_v2.preprocess_input
rescale = tf.keras.layers.experimental.preprocessing.Rescaling(1./127.5, offset= -1)
```

<h1>Building the Model</h1>

```python
# Create the base model from the pre-trained model MobileNet V2
img_shape = (img_height, img_width, 3)
base_model = tf.keras.applications.MobileNetV2(input_shape=img_shape,
                                               include_top=False,
                                               weights='imagenet')
```

```python
image_batch, label_batch = next(iter(train_dataset))
feature_batch = base_model(image_batch)
print(feature_batch.shape)
```

```python
base_model.trainable = False
```

```python
# Let's take a look at the base model architecture
base_model.summary()
```

```python
global_average_layer = tf.keras.layers.GlobalAveragePooling2D()
feature_batch_average = global_average_layer(feature_batch)
print(feature_batch_average.shape)
```

```python
prediction_layer = tf.keras.layers.Dense(1)
prediction_batch = prediction_layer(feature_batch_average)
print(prediction_batch.shape)
```

```python
inputs = tf.keras.Input(shape=(160, 160, 3))
x = data_augmentation(inputs)
x = preprocess_input(x)
x = base_model(x, training=False)
x = global_average_layer(x)
x = tf.keras.layers.Dropout(0.2)(x)
outputs = prediction_layer(x)
model = tf.keras.Model(inputs, outputs)
```

```python
base_learning_rate = 0.0001
model.compile(optimizer=tf.keras.optimizers.Adam(lr=base_learning_rate),
              loss=tf.keras.losses.BinaryCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

```python
model.summary()
```

```python
len(model.trainable_variables)
```

```python
initial_epochs = 10

loss0, accuracy0 = model.evaluate(validation_dataset)
```

```python
print("initial loss: {:.2f}".format(loss0))
print("initial accuracy: {:.2f}".format(accuracy0))
```

```python
history = model.fit(train_dataset,
                    epochs=initial_epochs,
                    validation_data=validation_dataset)
```

```python
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']

loss = history.history['loss']
val_loss = history.history['val_loss']

plt.figure(figsize=(8, 8))
plt.subplot(2, 1, 1)
plt.plot(acc, label='Training Accuracy')
plt.plot(val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.ylabel('Accuracy')
plt.ylim([min(plt.ylim()),1])
plt.title('Training and Validation Accuracy')

plt.subplot(2, 1, 2)
plt.plot(loss, label='Training Loss')
plt.plot(val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.ylabel('Cross Entropy')
plt.ylim([0,1.0])
plt.title('Training and Validation Loss')
plt.xlabel('epoch')
plt.show()
```

```python
loss, accuracy = model.evaluate(test_dataset)
print('Test accuracy :', accuracy)
```
